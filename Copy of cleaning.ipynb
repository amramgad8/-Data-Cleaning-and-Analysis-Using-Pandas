```python
df = df.drop_duplicates()
```

```python
# split up span into start and date /// expand=True       

df[['Start_Year','End_Year']] = df['Span'].str.split('-',expand=True)
```

```python
# drop column Span
df.drop('Span',axis=1,inplace=True)
```

```python
df.head(10)
```

```python
df['Player'].str.split('(',expand=True)
```

```python
df['Player'] = df['Player'].str.split('(',expand=True)[0]
```

```python
# df = df.drop('Country',axis=1)
df.head(10)
```

```python
# remove * from the number
df['Highest_Inns_Score'] = df['Highest_Inns_Score'].str.split('*',expand=True)[0]
```

```python
# removede ( - ) from coolumn
df = df[~(df['Balls_Faced'] == '-')]
```

```python
df.head(10)
```

```python
df['Balls_Faced'] = df['Balls_Faced'].str.split('+',expand=True)[0]
```

```python
df['4s'] = df['4s'].str.split('+',expand=True)[0]
```

```python
df['6s'] = df['6s'].str.split('+',expand=True)[0]
```

```python
df.tail(10)
```

```python
df['Matches'] = df['Matches'].str.split('*',expand=True)[0]
```

```python
df['Matches'] = df['Matches'].astype('int')
df['Highest_Inns_Score'] = df['Highest_Inns_Score'].astype('int')
df['Balls_Faced'] = df['Balls_Faced'].astype('int')
df['Batting_Strike_Rate'] = df['Batting_Strike_Rate'].astype('float')
df['4s'] = df['4s'].astype('int')
df['6s'] = df['6s'].astype('int')
df['Start_Year'] = df['Start_Year'].astype('int')
df['End_Year'] = df['End_Year'].astype('int')
```
